```python
import cv2
import os
from retinaface import RetinaFace
from deepface import DeepFace
from concurrent.futures import ThreadPoolExecutor
```

```python
def list_files_in_folder(folder_path):
    try:
        items = os.listdir(folder_path)
        files = []

        for item in items:
            item_path = os.path.join(folder_path, item)

            if os.path.isfile(item_path):
                files.append(item_path)
            elif os.path.isdir(item_path):
                subfolder_files = list_files_in_folder(item_path)
                files.extend(subfolder_files)

        return files

    except OSError as e:
        print("Error:", e)
        return []
```

```python
def list_folders_in_folder(folder_path):
    try:
        items = os.listdir(folder_path)
        files = []
        folders = [item for item in items if os.path.isdir(os.path.join(folder_path, item))]

        for folder in folders:
            folder_path_full = os.path.join(folder_path, folder)

            if not folder_path_full.endswith(os.path.sep):
                folder_path_full += os.path.sep

            files_in_subfolder = list_files_in_folder(folder_path_full)
            files.extend(files_in_subfolder)

        return files

    except OSError as e:
        print("Error:", e)
        return []
```

```python
def foldername(filename):
    name = filename.split("\\")

    return (name[len(name)-2])
```

```python
def process_frame(frame, database_files, sec):
    try:
        for file in database_files:
            result = DeepFace.verify(file, frame, model_name="ArcFace",
                                     detector_backend="retinaface", distance_metric="cosine",
                                     enforce_detection=False, align=True)
            if result['verified']:
              x = file.split("/")
              print(f"At second {sec}: \n" + str(x[len(x)-2]))

    except Exception as e:
        pass
```

```python
# Setting video name
videoname = "/content/drive/MyDrive/Facial Recognition/Samples/Top 10 Bollywood Dance Sequences of the Last Decade.mp4"

# Setting up video to capture frames
video = cv2.VideoCapture(videoname)

# Setting up a database of people to be identified.
database = "/content/drive/MyDrive/Facial Recognition/Database"

# Setting up the directory
# os.chdir("D:\\Academics\\Project\\Facial Recog\\temp")

# Setting up temp folder
tempfolder = "/content/drive/MyDrive/Facial Recognition/temp"

# Getting video properties.
fps = video.get(cv2.CAP_PROP_FPS)
frames = video.get(cv2.CAP_PROP_FRAME_COUNT)
seconds = round(frames / fps)
```

```python
sec = 10
counter = 0
count = 0
files = list_folders_in_folder(database)
```

```python
with ThreadPoolExecutor(max_workers=10) as executor:
    while sec < seconds:
        t_msec = int(1000 * sec)
        video.set(cv2.CAP_PROP_POS_MSEC, t_msec)
        ret, frame = video.read()
        filename1 = f"/content/drive/MyDrive/Facial Recognition/temp/image_{count}.jpg"
        cv2.imwrite(filename1, frame)

        try:
            img = cv2.imread(filename1)
            executor.submit(process_frame, filename1, files, sec)
            count = count + 1
            sec += 1
        except Exception as e:
            sec += 1

# Release the video capture object
video.release()

# Close all OpenCV windows
cv2.destroyAllWindows()
```
