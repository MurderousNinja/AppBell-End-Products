```python
url1 = 'https://drive.google.com/uc?id={}'.format(file_id1)
url2 = 'https://drive.google.com/uc?id={}'.format(file_id2)
```

```python
from google.colab.patches import cv2_imshow

# read image
img = cv2.imread(url1)

# Display the image
cv2.imshow("Image", img)

# Wait for the user to press a key
cv2.waitKey(0)

# Close all windows
cv2.destroyAllWindows()
```
